<h5 style="color:Red">Задание 1.</h5><a id='task1'></a> - Напишите функцию, которая классифицирует фильмы по следующим правилам:
- оценка 2 и меньше - низкий рейтинг  
- оценка 4 и меньше - средний рейтинг  
- оценка 4.5 и 5 - высокий рейтинг  

Результат классификации запишите в столбец class

```python
import pandas as pd 
```

```python
ratings = pd.read_csv('ratings.csv', encoding = 'UTF-8') #открываем таблицу с рейтингами
ratings.head()
```

```python
movies = pd.read_csv('movies.csv', encoding = 'UTF-8') #открываем таблицу с фильмами 
movies.head()
```

```python
merged = ratings.merge(movies[['movieId', 'title']], on = 'movieId', how = 'left') #объединяем по идентификатору фильма
```

```python
average = merged[['movieId','title', 'rating']].groupby(['movieId', 'title']).mean().round(decimals = 2) #определяем среднее значение рейтинга
average.head()
```

```python
#функция для классификации рейтинга
def movie_rating(rating):
    if rating <= 2:
        return 'Низкий рейтинг'
    elif rating <= 4:
        return 'Средний рейтинг'
    elif rating == 4.5 or rating == 5.0:
        return 'Высокий рейтинг'
```

```python
average['class'] = average['rating'].apply(movie_rating) #создаем столбец класс и применяем функцию
average.tail()
```

<h5 style="color:Red">Задание 2.</h5><a id='task2'></a> - Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

```python
geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
```

```python
keywords = pd.read_csv('keywords.csv', encoding = 'UTF-8')
keywords.head()
```

```python
def classificator(keyword):
    for reg, value in geo_data.items():
        for city in value:
            if city in keyword:
                return reg
    return 'undefined'
```

```python
keywords['region'] = keywords['keyword'].apply(classificator)
keywords.head(10) 
```
